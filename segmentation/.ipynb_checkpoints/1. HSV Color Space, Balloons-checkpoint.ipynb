# HSV Color Space, Balloons

### Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
%matplotlib inline

# Read in the image
image = cv2.imread('a1.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

### Plot color channels

```python
# RGB channels
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]
x= image[:,:,2]

f, (ax1, ax2, ax3,ax4) = plt.subplots(1, 4, figsize=(20,10))

ax1.set_title('Red')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b, cmap='gray')
ax4.set_title('Bluex')
ax4.imshow(x, cmap='gray')

```

```python
# Convert from RGB to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# HSV channels
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

### Define pink and hue selection thresholds

```python
# Define our color selection criteria in HSV values
#lower_hue = np.array([160,0,0]) 
#upper_hue = np.array([180,255,255])
```

```python
lower_hue = np.array([160,0,0]) 
upper_hue = np.array([180,255,255])
```

```python
# Define our color selection criteria in RGB values
lower_pink = np.array([180,0,100]) 
upper_pink = np.array([255,255,230])
```

```python

lower_green = np.array([50,60,60])
upper_green = np.array([100,255,255])
```

### Mask the image 

```python
# Define the masked area in RGB space
mask_rgb = cv2.inRange(image, lower_pink, upper_pink)

# mask the image
masked_image = np.copy(image)
masked_image[mask_rgb==0] = [0,0,0]

# Vizualize the mask
plt.imshow(masked_image)
```

```python
# Now try HSV!

# Define the masked area in HSV space
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)

# mask the image
masked_image = np.copy(image)
masked_image[mask_hsv !=0] = [0,255,0]

# Vizualize the mask
plt.imshow(masked_image)
```

```python
# Now try HSV!

# Define the masked area in HSV space
mask_hsv = cv2.inRange(hsv, lower_green, upper_green)

# mask the image
masked_image = np.copy(image)
masked_image[mask_hsv !=0] = [0,255,0]

# Vizualize the mask
plt.imshow(masked_image)
```

```python
output=image - masked_image
```
